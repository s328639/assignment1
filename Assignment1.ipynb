```python
# import of lib and datafile
import pandas as pd
import matplotlib
import seaborn as sns
from matplotlib import pyplot as plt

dataframe = pd.read_csv("Ruter_data.csv", delimiter=";")

dataframe.shape
```

```python
#Linje i Oslo med flest avganger
```

```python
line_with_most_departures_oslo = dataframe[dataframe["Linjefylke"] == "Oslo"]
```

```python
line_with_most_departures_oslo['Linjenavn'].value_counts().plot.barh()
```

```python
#holdeplass i Oslo med flest avganger
```

```python
holdeplass_most_departures = dataframe[dataframe["Fylke"] == "Oslo"]
```

```python
plt.figure(figsize=(10,5))
chart = sns.countplot(
    data=holdeplass_most_departures,
    x='Holdeplass_Fra',
    order=pd.value_counts(holdeplass_most_departures['Holdeplass_Fra']).iloc[:10].index
)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
```

```python
#hvilke holdeplasser har flest påstigende passasjerer
```

```python
onboardmax = dataframe[dataframe["Passasjerer_Ombord"] >= 40]
```

```python
fix, ax = plt.subplots()
data = onboardmax
Fra_Holdeplass = onboardmax.Holdeplass_Fra
Antall_pastigende = onboardmax.Passasjerer_Ombord
ax.bar(Fra_Holdeplass,Antall_pastigende,width=0.4)
ax.set_title('Holdeplasser med flest påstigende passasjerer')
ax.set_xlabel('Fra holdeplass')
ax.set_ylabel('Påstigende passasjerer')
for tick in ax.get_xticklabels():
    tick.set_rotation(90)
```

```python
#Hvilke holdeplasser har flest avstigende passasjerer
```

```python
offboardmax = dataframe[dataframe["Passasjerer_Ombord"] <= -22]
```

```python
fix, ax = plt.subplots()
data = offboardmax
Fra_Holdeplass = offboardmax.Holdeplass_Fra
Antall_pastigende = offboardmax.Passasjerer_Ombord
ax.bar(Fra_Holdeplass,Antall_pastigende,width=0.4)
ax.set_title('Holdeplasser med flest avstigende passasjerer')
ax.set_xlabel('Fra holdeplass')
ax.set_ylabel('avstigende passasjerer')
for tick in ax.get_xticklabels():
    tick.set_rotation(90)
```
